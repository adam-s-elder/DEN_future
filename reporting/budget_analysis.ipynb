```python
# group by year/by state, sort by amount
# Group by 'state'

grouped = full_df.groupby(['State/Territory', pd.to_datetime(full_df['timestamp']).dt.year])

# Example: Calculate the total population and total area for each state
result = grouped.agg({'Total_Cumulative_Expenditures': 'sum'}).reset_index()
result = result.sort_values(by='Total_Cumulative_Expenditures', ascending=False)

display(result)
```

```python
nlp = spacy.load("en_core_web_sm")

# Function to extract keywords
def extract_keywords(text):
    doc = nlp(text)
    key_words = set()
    for token in doc:
        if token.pos_ in ["NOUN", "PROPN", "ADJ", "VERB"] and not token.is_stop:
            key_words.add(token.text)
    return list(sorted(set(key_words)))

# Apply the function to the 'text' column and create a new 'keywords' column
full_df['Project_Description_keywords'] = full_df['Project_Description'].apply(extract_keywords)

def check_trace(text):
    if 'trace' in text or 'tracing' in text:
        return 'yes'
    else:
        return 'no'
        
full_df['Project_mention_tracing_yn'] = full_df['Project_Description_keywords'].apply(lambda x: check_trace(x))
```

```python
display(Markdown('### Sample of Project Descriptions'))
```

```python
percentage = (len(full_df[full_df['Project_mention_tracing_yn'] == 'yes']) / len(full_df)) * 100

# Display the result as Markdown
display(Markdown(f'#### {percentage:.2f}% of descriptions mention tracing'))
```

```python
# do something with Project Description
for index, row in full_df[0:15].iterrows():
    print(f'Item #{index}')
    print(f'Recipient_Type: {row.Recipient_Type}')
    print(f'Project_Name: {row.Project_Name}')
    print(f'Project_Name: {row.Recipient_Name}', '\n')
    print(f'Mentions tracing?: {row.Project_mention_tracing_yn}')
    print(row.Project_Description)
    print(f'Keywords: {row.Project_Description_keywords}', '\n')
```
